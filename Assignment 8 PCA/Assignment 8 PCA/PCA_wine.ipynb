```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
data=pd.read_csv('/content/wine.csv')
data
```

```python
data.shape
```

```python
# no null values
data.info()
```

```python
data['Type'].value_counts()
```

```python

wine=data.iloc[:,1:]
wine
```

```python
# Converting data to numpy array
wine_array=wine.values
wine_array
```

```python
#Normalizaing the numerical data
wine_norm=scale(wine_array)
wine_norm
```

# PCA Implementation

```python
# applying PCA Fit Transform to dataset
pca=PCA(n_components=13)

wine_pca=pca.fit_transform(wine_norm)
wine_pca
```

```python
# components matrix
pca.components_
```

```python
#the amount of varience that each pca has 
var=pca.explained_variance_ratio_
var
```

```python
# Cummulative varience of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
#varienece plot
plt.plot(var1,color='Green')
```

```python
# Final Dataframe
final_df=pd.concat([data['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
#visulization
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

# Checking cluster with other alogorithms


##  1.Hierarchical Clustring

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# create dendrogram
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
# create cluster
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python

y=pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding cluster to dataset
wine2=wine.copy()
wine2['clustersid']=hclusters.labels_
wine2
```

## 2. K=Means Clustring

```python
from sklearn.cluster import KMeans
```

```python
# data is already normalize data 
```

```python
# within cluster sum of square
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
# plot k value range vs wcss
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Building cluster algorithm using K=3

```python
# Cluster algorithm using K=3
clusters3=KMeans(3,random_state=30).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to the data
wine3=wine.copy()
wine3['cluster3id']=cluster3.labels_
wine3
```
