## Model Building

```python
# Linear model 
import statsmodels.formula.api as smf
```

```python
# Linear model 
linear_model=smf.ols('Sales~No',data=Train).fit()
pred_linear=pd.Series(linear_model.predict(pd.DataFrame(Test['No'])))
rmse_linear=np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_linear))**2))
rmse_linear
```

```python
# Exponential model
expo_model=smf.ols('log_sales~No',data=Train).fit()
pred_expo=pd.Series(expo_model.predict(pd.DataFrame(Test['No'])))
rmse_expo=np.sqrt(np.mean((np.array(Test['Sales'])-np.exp(np.array(pred_expo)))**2))
rmse_expo
```

```python
# Quadratic Model
quad_model=smf.ols('Sales~No+No_squared',data=Train).fit()
pred_quad = pd.Series(quad_model.predict(Test[['No','No_squared']]))
rmse_quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_quad))**2))
rmse_quad
```

```python
# Additive Seasonality
add_sea = smf.ols('Sales~Q1+Q2+Q3+Q4',data=Train).fit()
pred_add_sea = pd.Series(add_sea.predict(Test[['Q1','Q2','Q3','Q4']]))
rmse_add_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea))**2))
rmse_add_sea
```

```python
# Additive Seasonality Quadratic
add_sea_quad = smf.ols('Sales~No+No_squared+Q1+Q2+Q3+Q4',data=Train).fit()
pred_add_sea_quad = pd.Series(add_sea_quad.predict(Test[['No','No_squared','Q1','Q2','Q3','Q4']]))
rmse_add_sea_quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad
```

```python
# Muiltiplicative Seasonality
Multi_sea = smf.ols('log_sales~Q1+Q2+Q3+Q4',data=Train).fit()
pred_Multi_sea = pd.Series(Multi_sea.predict(Test))
rmse_Multi_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Multi_sea)))**2))
rmse_Multi_sea
```

```python
# Muiltiplicative Additive Seasonality
Multi_ADD_sea = smf.ols('log_sales~No+Q1+Q2+Q3+Q4',data=Train).fit()
pred_Multi_ADD_sea = pd.Series(Multi_ADD_sea.predict(Test))
rmse_Multi_ADD_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Multi_ADD_sea)))**2))
rmse_Multi_ADD_sea
```

```python
df2={'Model':pd.Series(['linear_model','expo_model','quad_model','add_sea','add_sea_quad','Multi_sea','Multi_ADD_sea']),
    'RMSE_Values':pd.Series([rmse_linear,rmse_expo,rmse_quad,rmse_add_sea,rmse_add_sea_quad,rmse_Multi_sea,rmse_Multi_ADD_sea,])}
Table_rmse=pd.DataFrame(df2)
Table_rmse
```
